```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
rig_count = pd.read_excel('North America Rotary Rig Count (Jan 2000 - Current) (2).xlsx',
                         sheet_name=1, parse_dates=True, skiprows=8, header=[1,2])
```

```python
rig_count.head()
```

```python
rig_count.columns = [f'{i} {j}' for i, j in rig_count.columns]
```

```python
rig_count.head()
```

```python
rig_count.columns
```

```python
rig_count_oil=rig_count.loc[:,rig_count.columns.str.contains('Oil')]
```

```python
pd.set_option('mode.chained_assignment', None)
rig_count_oil['Date'] = rig_count['Unnamed: 0_level_0 Date']
```

```python
#rig_count_oil.set_index('Date', inplace=True)
rig_count_oil.head()
```

```python
print(rig_count_oil.info())
```

```python
#Drop data from gas-weighted basins
rig_count_oil_oil_basins = rig_count_oil.drop(['Fayetteville Oil', 'Haynesville Oil', 'Marcellus Oil', 'Utica Oil', 'Others Oil'], axis=1)
```

```python
plt.figure(figsize=(15,7))
plt.style.use('default')
palette=plt.get_cmap('Dark2')
num=0
for column in rig_count_oil_oil_basins.drop('Total US RigCount Oil', axis=1):
    num+=1
    plt.plot(rig_count_oil_oil_basins.index, rig_count_oil_oil_basins[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)
plt.legend()
```

```python
oil_price = pd.read_excel('RCLC1w.xls', sheet_name=1, header=2, parse_dates=True)
oil_price.head()
```

```python
oil_price_post2010 = oil_price[oil_price['Date'].dt.year>2010]
```

```python
oil_price_post2010.rename(columns={'Weekly Cushing, OK Crude Oil Future Contract 1  (Dollars per Barrel)': 'Oil Price'}, inplace=True)
oil_price_post2010.head()
```

```python
oil_price_post2010 = oil_price_post2010.iloc[4:,:]
```

```python
print(oil_price_post2010.shape)
print(rig_count_oil_oil_basins.shape)
```

```python
rig_count_oil_oil_basins = rig_count_oil_oil_basins.iloc[:-1,:]
```

```python
widths = [6, 6, 6]
heights = [6, 2.5, 6, 2.5, 6, 2.5]
gs_kw = dict(width_ratios=widths, height_ratios=heights)
fig1, f1_axes = plt.subplots(ncols=3, nrows=6, gridspec_kw=gs_kw, figsize=(12,12))
#for r, row in enumerate(f1_axes):
#    for c, ax in enumerate(row):
#        label = 'Width: {}\nHeight: {}'.format(widths[c], heights[r])
#        ax.annotate(label, (0.1, 0.5), xycoords='axes fraction', va='center')
```

```python
plt.style.use('seaborn-colorblind')
#palette=sns.color_palette('Blues_d', 10, desat=.5)
fig, axs = plt.subplots(2,1, figsize=(14,8), sharex=True)
labels = rig_count_oil_oil_basins.drop('Total US RigCount Oil', axis=1).columns
axs[0].stackplot(rig_count_oil_oil_basins.index, rig_count_oil_oil_basins.drop('Total US RigCount Oil', axis=1).T, alpha=0.6, labels=labels)
axs[1].plot(oil_price_post2010['Date'], oil_price_post2010['Oil Price'], alpha=0.7)
plt.tick_params(bottom=False, left=False)
for ax in axs:
    for a in ax.spines:
        ax.spines[a].set_visible(False)
axs[0].legend(ncol=2, frameon=False)
plt.subplots_adjust(hspace=0.07)
#plt.margins(x=.05)
```

```python
plt.rcParams['savefig.facecolor']='White'
plt.style.use('seaborn-colorblind')
#palette=sns.color_palette('Blues_d', 10, desat=.5)
widths = [1]
heights = [3,1]
gs_kw = dict(width_ratios=widths, height_ratios=heights)
fig1, fig1_axes = plt.subplots(ncols=1, nrows=2, figsize=(14,8), gridspec_kw=gs_kw, sharex=True)
labels = rig_count_oil_oil_basins.drop('Total US RigCount Oil', axis=1).columns
fig1_axes[0].stackplot(rig_count_oil_oil_basins.index, rig_count_oil_oil_basins.drop('Total US RigCount Oil', axis=1).T, alpha=0.6, labels=labels)
fig1_axes[1].plot(oil_price_post2010['Date'], oil_price_post2010['Oil Price'], alpha=0.7)
plt.tick_params(bottom=False, left=False)
for ax in fig1_axes:
    for a in ax.spines:
        ax.spines[a].set_visible(False)
fig1_axes[0].legend(ncol=2, frameon=False)
fig1_axes[0].set_ylabel('# of Rigs')
fig1_axes[1].set_ylabel('Oil Price (USD)')
fig1_axes[0].set_title('U.S. Onshore Oil Rig Count Response to Price of Oil (2011-Current)', fontsize='x-large', fontweight='bold')
fig1_axes[0].annotate('Williston Basin', (0.25,0.78), xycoords='figure fraction', fontweight='bold')
fig1_axes[0].annotate('Permian Basin', (0.25,0.62), xycoords='figure fraction', fontweight='bold')
fig1_axes[0].annotate('Eagle Ford', (0.25,0.37), xycoords='figure fraction', fontweight='bold')
plt.subplots_adjust(hspace=0.07)
plt.tight_layout()
plt.savefig('Oil Rig Count vs Oil Price.png')
```
